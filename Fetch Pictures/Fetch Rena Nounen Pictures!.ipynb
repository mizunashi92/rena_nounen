```python
import pandas as pd
a = pd.read_csv("non.csv")
```

```python
a.shape
a.columns = ['Name']
```

```python
a
```

```python
get = []
for non in a["Name"]:
    if "imgc" in non:
        get.append(non)
        
```

```python
l = []
for love in get:
    c = love.split("//")
    d = c[1].split("&t")
    l.append(d[0])
```

```python
import urllib.request

for sintyadi, rena in enumerate(l):
    try:
        http = "https://"+rena
        file = "mylove_"+str(sintyadi)+".jpg"
        print("reading", http)
        urllib.request.urlretrieve(http, file)
    except:
        nounen = rena.split("&amp")
        http = "https://"+nounen[0]
        file = "mylove_"+str(sintyadi)+".jpg"
        print("reading", http)
        urllib.request.urlretrieve(http, file)
    
```
